```python
import connections
import numpy as np
import util
```

```python
def check_forward_history():
    rec_conn = connections.RecurrentConnect(5, 3)
    input_node1 = util.create_np_node(3, 5, init_random=True)
    rec_conn.set_input_nodes([input_node1])
    for i in xrange(7):
        print 'round-', i, '; inputdata: ', rec_conn._input_nodes[0].get_data()
        rec_conn.forward()
        print 'round-', i, '; outputdata: ', rec_conn._output_nodes[0].get_data()
        for j in xrange(rec_conn._hist_index):
            print '\thist-', j, '; histdata:', rec_conn._hist_nodes[j].get_data()
        

check_forward_history()
```

```python
def check_back_prob():
    rec_conn = connections.RecurrentConnect(5, 3)
    input_node1 = util.create_np_node(3, 5, init_random=True)
    rec_conn.set_input_nodes([input_node1])
    rec_conn.forward()
    grad = rec_conn.get_output_nodes()[0].get_grad()
    grad -= 0.5
    rec_conn.backprob()
    print "hist_prob: ", rec_conn._hist_nodes[0].get_grad()
    print "input_prob 0: ", rec_conn._input_nodes[0].get_grad()
    print "input_prob 1: ", rec_conn._input_nodes[1].get_grad()

check_back_prob()
```

```python
def check_hist_update():
    rec_conn = connections.RecurrentConnect(5, 3)
    input_node1 = util.create_np_node(3, 5, init_random=True)
    rec_conn.set_input_nodes([input_node1])
    for i in xrange(5):
        rec_conn.forward()
    grad = rec_conn.get_output_nodes()[0].get_grad()
    grad -= 0.5
    rec_conn.backprob()
    
    print "hist_prob: ", rec_conn._hist_nodes[0].get_grad()
    print "input_prob 0: ", rec_conn._input_nodes[0].get_grad()
    print "input_prob 1: ", rec_conn._input_nodes[1].get_grad()
    
    for i in xrange(rec_conn._hist_index-1):
        input_data = rec_conn._hist_nodes[i+1].get_data()
        gradient_data = rec_conn._hist_nodes[i].get_grad()
        print "round-", i, "; input_data:", input_data
        print "round-", i, "; gradient_data:", gradient_data
        rec_conn._updator.cal_update_values(input_data, gradient_data)
        rec_conn._updator.do_update(rec_conn._layer.get_weights())
        #cal new gradient
        c_input_nodes = [rec_conn._hist_nodes[i+1]]
        c_output_nodes = [rec_conn._hist_nodes[i], rec_conn._output_nodes[1]]
        rec_conn._layer.backprob(c_input_nodes, c_output_nodes)

check_hist_update()
```
